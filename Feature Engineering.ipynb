```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler , MinMaxScaler
```

```python
df = pd.read_csv(r'C:\Users\123\Desktop\covid_toy.csv')
```

```python
df.head()
```

```python
df=df.fillna(df['fever'].mean())
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

1.Aam Zindagi

```python
#ordinalencoding --> cough
oe = OrdinalEncoder(categories=[['Mild','Strong']])
x_train_cough = oe.fit_transform(x_train[['cough']])

# also the test data
x_test_cough = oe.fit_transform(x_test[['cough']])

x_train_cough.shape
```

```python
# OneHotEncoder ---> gender,city
ohe = OneHotEncoder(drop='first',sparse_output=False)
x_train_gender_city = ohe.fit_transform(x_train[['gender','city']])

# also the test data
x_test_gender_city = ohe.fit_transform(x_test[['gender','city']])

x_train_gender_city.shape
```

```python
# Extracting Age
x_train_age = x_train.drop(columns=['gender','cough','city']).values

# also the test data
x_test_age = x_test.drop(columns=['gender','cough','city']).values

x_train_age.shape
```

```python
x_train_transformed = np.concatenate((x_train_age,x_train_gender_city,x_train_cough),axis=1)

# also the test data
x_test_transformed = np.concatenate((x_test_age,x_test_gender_city,x_test_cough),axis=1)

x_train_transformed.shape
```

Mentos Zindagi

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf2',OneHotEncoder(drop='first',sparse_output=False),['gender','city'])   
],remainder='passthrough')
```

```python
x_train_transformed = transformer.fit_transform(x_train).shape
print(x_train_transformed)
```

```python
x_test_transformed = transformer.transform(x_test)
print(x_test_transformed)
```
