Super method:

super() method is used to access methods of the parent class.

```python
class Car:
    def __init__(self,type):
        self.type = type

    @staticmethod                      # Super method
    def start():
        print("car started..")

    @staticmethod
    def stop():
        print("car stopped.")
        
class ToyotaCar(Car):
    def __init__(self,name,type):
        super().__init__(type)
        self.name = name
        super().start()
        
car1 = ToyotaCar("civic","electric")
print(car1.type)
```

```python
# MUltiple Inheritance:
# when a derived class uses more than one base class it is said to be multiple Inheritance
class Music_system:
    def __init__(self,speakers):
        self.speakers = speakers

    def music_details_system(self):
        print("This is a class for music system.")
        
class Engine:
    def __init__(self,engine):
        self.engine = engine

    def engine_details(self):
        print("This is a class for car engine.")

class Car(Music_system,  Engine):
    def __init__(self, speakers, engine, model):
        Music_system.__init__(self, speakers)
        Engine.__init__(self, engine)
        self.model = model

    def car_details(self):
        print("This is main car.")
        
fortuner = Car(7,"petrol","fortuner")
print(fortuner.speakers)
print(fortuner.music_details_system())
print(fortuner.engine)
print(fortuner.engine_details())
print(fortuner.model)
print(fortuner.car_details())
```

```python
class Shape_old:
     def area(self):
        print("welcome to area calculator old.")

class Shape(Shape_old):
     def __init__(self,length,width):
        self.length = length
        self.width = width 

    
class Rectangle(Shape):
    def __init__(self,length,width):
          super().__init__(length,width)
    def area(self):                 
        super().area()                       
        print("Area of Rectangle is =",self.length*self.width)

class Cube(Shape):
    def __init__(self, length, width, height):
        super().__init__(length,width)
        self.height = height
    def area(self):
        super().area()
        print("Area of cube is =",self.length*self.width*self.height)

rect = Rectangle(5,6)
rect.area()

cube = Cube(5,6,10)
cube.area()
    
```
