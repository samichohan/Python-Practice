USER DEFINED EXCEPTIONS:

```python
class InvalidAgeError(Exception):
    pass
class AgeMoreError(Exception):
    def __init__(self,age):
        self.age = age

    def error_message(self):
        print("You are not allowed ,Your age is =", self.age)

age = 62
try:
    if age < 3:
        raise AgeLessError
    elif age > 60:
        raise AgeMoreError(age)
    else:
        print("you are allowed")
except AgeLessError:
    print("Your age is less than required")
except AgeMoreError as e:
    e.error_message()
    pass
    
```
