1.Matplotlip Bar Plot:

```python
x=['python','c','c++','Java']
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel('language',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.title(' Bar Graph',fontsize=15)
c=['y','b','m','g']
plt.bar(x,y,width=0.4,color='blue',label='Popularity')
plt.bar(x,z,width=0.4,color='red',label='Popularity1')
plt.legend()
plt.show()
```

```python
import numpy as np
x=['python','c','c++','Java']
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('language',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.title(' Bar Graph',fontsize=15)

plt.bar(p,y,width,color='blue',label='Popularity')
plt.bar(p1,z,width,color='red',label='Popularity1')

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
x=['python','c','c++','Java']
y=[85,70,60,82]
z=[20,30,40,50]

plt.xlabel('language',fontsize=15)
plt.ylabel('No.',fontsize=15)
plt.title(' Bar Graph',fontsize=15)

plt.barh(x,y,color='blue',label='Popularity')
plt.barh(x,z,color='red',label='Popularity1')

plt.legend()
plt.show()
```

2.scatter Plot:

```python
import matplotlib.pyplot as plt
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]

plt.scatter(day,no,color='red')
plt.title('Scatter Plot',fontsize=15)
plt.xlabel('Day',fontsize=13)
plt.ylabel('No.',fontsize=13)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no2=[3,2,4,5,1,6,2]
colors=[10,49,30,29,56,20,30]
sizes=[400,200,300,500,600,200,400]
plt.scatter(day,no,c=colors,s=sizes,cmap='viridis',alpha=0.5)
plt.scatter(day,no2,color='red',s=sizes,alpha=0.5)
t=plt.colorbar()
t.set_label('colorbar',fontsize=15)
plt.title('Scatter Plot',fontsize=15)
plt.xlabel('Day',fontsize=13)
plt.ylabel('No.',fontsize=13)
plt.show()
```

4.Pie plot:

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=['c','c++','Java','python']
ex=[0.3,0.0,0.0,0.0]
c=['r','b','g','y']
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',shadow=True,radius=1,labeldistance=1.1,startangle=0
        ,textprops={'fontsize':15},counterclock=True,wedgeprops={'linewidth':1,'edgecolor':'w'}
       ,rotatelabels=False)
plt.title('AI & DATASCIENCE')
plt.legend()
plt.show()
```
