TYPES OF ARGUMENTS:
In Python, we have the following types of Arguments:

1. Positional Arguments
2. Keyword Arguments
3. Default Arguments
4. Arbitrary Arguments

```python
# positional Arguments :
# positional arguments are the arguments that need to be passed in the correct 
# predifined order.

def greater(a, b, c):
    if (a > b and a > c):
        print("First is greater")
    elif (b > c and b > a):
        print("Second is greater")
    else:
        print("Third is greater")

x = 2
y = 3
z = 4

greater(x,z,y)
```

```python
# Keyword Arguments: 
# keyword Arguments are the arguments in which parameters name with equal to sign is used for
# Keyword arguments, we need not maintain the arguments.

def greater(a, b, c):
    if (a > b and a > c):
        print("First is greater")
    elif (b > c and b > a):
        print("Second is greater")
    else:
        print("Third is greater")

x = 2
y = 3
z = 4

greater(a = x,c = z,b = y)
```

```python
# default  Arguments:
# We can set some values of the arguments while defining the function itself .
# these values are known as default values.

def greater(a, b, c = 99):
    print(a,b,c)
              
x = 2
y = 3
z = 4

greater(a = x,b = y,c = z)

def greater(a, b, c = 99):
    print("a =",a)
    print("b =",b)
    print("c =",c)          


x = 2
y = 3
z = 4

greater(a = x,b = y)
```

```python
# Arbirary Arguments : Variable length Arguments:
# 1. *args (variable length positional arguments):
# when we don't know the numbers of arguments we want to pass the functions , we can use on asterisk
# (*) symbol before the parameters name in the functions All the arguments will be passwd
# as a tuple.

def greater(*args):
    print("args =",args)
    print(args[1])

x = 2
y = 3
z = 4

greater(x,y,z,5,6,7)


def greater(*args):
    print(sum(args)/len(args))

x = 2
y = 3
z = 4

greater(x,y,z,5,6,7,8,9)
```

```python
# **kwargs(variables length keyword arguments): dictionary
def greater(**kwargs):
    print("kwargs =",kwargs)

x = 2
y = 3
z = 4

greater(x = x,y = y,z = z,a = z,b = 5) 
```

```python
def greater(**kwargs):
    print("kwargs =",kwargs)
    print("x =",kwargs['x'])
    print("b =",kwargs['b'])

x = 2
y = 3
z = 4

greater(y = y,z = z,a = z,b = 5,x = x) 
```

```python
# map --> the python map() functions applied the specified functions is each items of an iterable like a list,tuple.
def square(n):
    return(n**2)

numbers = [1,2,3,4,5] 
numbers2 = map(square,numbers)
print(list(numbers2)) # tuple bhi kar sakte hai
```

```python
# map 
def square(n):
    return(n**2)

numbers = [1,2,3,4,5] 
numbers2 = map(square,numbers)
for i in numbers2:
    print(i)
```

```python
# filter functions ---> the filter applies the specifi=ed functins to all the elements of the sequence aqnd return
# the elements for which the functions remains true,
def even(n):
    if n%2 == 0:
        return True
    else:
        return False
numbers = [1,2,3,4,5]

num3 = filter(even,numbers)
print(list(num3))
 
```

```python
# Reduce Functions : the reduce() function will perform the rolling computation on the sequential
# pair of the values It doews not return th sequence of values, but it return a single values at output
from functools import reduce
def add(a,b):
    return a + b

numbers = [1,2,3,4,5]
result = reduce(add,numbers)
print(result)
```
